```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Loan_Data.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include="float64").columns
```

```python
# using sklearn library we fill the missing value
from sklearn.impute import SimpleImputer

si = SimpleImputer(strategy= 'mean')
ar = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term','Credit_History']])
```

```python
new_dataset = pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```

```python
new_dataset.isnull().sum()
```
